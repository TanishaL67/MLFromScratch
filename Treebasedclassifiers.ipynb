First I am going to try a decision tree classifier.

A Decision Tree Classifier is a machine learning model used for classification or regression tasks, where data is split according to certain criteria (features and thresholds) to make predictions.

Decision trees are high-variance models if we keep brewing our trees every element will have its regions. Hence these trees require regularisation.
This can be done by:
- when reach a minimum leaf size stop splitting the leaf
- enforce maximun depths
- max number of nodes
- minimum decrease in loss (not good to use)
- pruning - grow the whole tree and then remove the leaf (use misclassification values with validation sets)

When there is a relation between the variables (they are not fully independent of each other) then we need to start with broader questions and less decrease in misclassification loss and than we reach the optimal questions and the decrease in loss is significant then.


Imagine you have n examples, f features, and d depth of your tree. Then the run time for the test is O(d) and train time each point is part of O(d) nodes and cost of point at each node is O(f), so the total cost will be O(dnf).


If we talk about downsides than it has a couple of them:
- it has no additive structure - will have problems when features are interacting additively with one another, it has to ask a lot of questions to reach an optimum decision boundary.
- Have very high variance
- have fairly low predictive accuracy

### Ensemble Methods

Decision trees can be improved a lot using ensemble models.

Ensembling - it works because the variance or the mean variance will decrease when we increase the number of random variables.
This is the basic work:
( x1, x2 .... ) set of n random variables independently identically distributed and xi has variance v^2 then the
var(1/n (x1+x2+...)) = v^2 / n (it is variance of mean)

- this independence assumption is not always correct so we drop it.


There are different ways to ensemble.
- Use different algorithms
- use different training sets
- Bagging - Random forest
- Boosting - Adaboost, XGBoost


Bagging - Bootstrap aggregation
Say you have true population p, and the training set s sample from p. Here assume your population is the training set i.e. p = s. Here you sample n times from s and do it with replacement and train your model on that. This is boostraping.  

Bootstrapping drives down correlation

We need to aggregate this by taking the bootstrap sample and aggregating the
So we sample more and more models that decrease the variance and we stop when the error slows down in decreasing. This slightly increases the bais because of random subsampling but generally, the decrease in variance is greater than the decrease. so this is a good tradeoff.


Boosting - It is more additive, we decrease the bias of the sample. It reweights the examples and adds them as you go on.
